```python
import pandas as pd
```

```python
mortality_df = pd.read_csv('original-data/mort.csv')

# Drop rows with NaN FIPS ID
mortality_df = mortality_df.dropna(subset=['FIPS'])
mortality_df['FIPS'] = mortality_df['FIPS'].astype(int)

mortality_df.head()
```

```python
aqi_df = pd.read_excel('original-data/ctyfactbook2023.xlsx', sheet_name='County Factbook 2023', skiprows=2)

# Drop rows with NaN FIPS ID
aqi_df = aqi_df.dropna(subset=['County FIPS Code'])
aqi_df['County FIPS Code'] = aqi_df['County FIPS Code'].astype(int)

aqi_df
```

```python
merged_df = pd.merge(mortality_df, aqi_df, left_on='FIPS', right_on='County FIPS Code', how='inner')

merged_df
```

```python
drop_columns = [
    'Location',
    'County FIPS Code',
    'State',
    'County',
    'Mortality Rate, 1980* (Min)',
    'Mortality Rate, 1980* (Max)',
    'Mortality Rate, 1985* (Min)',
    'Mortality Rate, 1985* (Max)',
    'Mortality Rate, 1990* (Min)',
    'Mortality Rate, 1990* (Max)',
    'Mortality Rate, 1995* (Min)',
    'Mortality Rate, 1995* (Max)',
    'Mortality Rate, 2000* (Min)',
    'Mortality Rate, 2000* (Max)',
    'Mortality Rate, 2005* (Min)',
    'Mortality Rate, 2005* (Max)',
    'Mortality Rate, 2010* (Min)',
    'Mortality Rate, 2010* (Max)',
    'Mortality Rate, 2014* (Min)',
    'Mortality Rate, 2014* (Max)',
    '% Change in Mortality Rate, 1980-2014',
    '% Change in Mortality Rate, 1980-2014 (Min)',
    '% Change in Mortality Rate, 1980-2014 (Max)'
]

trimmed_df = merged_df.drop(drop_columns, axis=1)

trimmed_df = trimmed_df.rename(columns={
    "CO          8-hr (ppm)": "CO 8-hr (ppm)",
    "Pb           3-mo (Âµg/m3)": "Pb 3-mo (mu g/m3)",
    "NO2         AM (ppb)": "NO2 AM (ppb)",
    "NO2          1-hr (ppb)": "NO2 1-hr (ppb)",
    "O3            8-hr (ppm)": "O3 8-hr (ppm)",
    "PM10        24-hr (Âµg/m3) ": "PM10 24-hr (mu g/m3)",
    "PM2.5     Wtd AM (Âµg/m3) ": "PM2.5 Wtd AM (mu g/m3)",
    "PM2.5     24-hr (Âµg/m3) ": "PM2.5 24-hr (mu g/m3)",
    "SO2         1-hr (ppb)": "SO2 1-hr (ppb)"
})

print(trimmed_df.columns.tolist())
trimmed_df
```

## Aggregate Data by Metropolitan Area

```python
cbsa_crosswalk = pd.read_csv('original-data/cbsa2fipsxw.csv', dtype={'fipsstatecode': str, 'fipscountycode': str})
cbsa_crosswalk['FIPS'] = cbsa_crosswalk['fipsstatecode'].astype(str) + cbsa_crosswalk['fipscountycode'].astype(str)
cbsa_crosswalk['FIPS'] = cbsa_crosswalk['FIPS'].astype(int)
cbsa_crosswalk.head()
```

```python
metro_df = trimmed_df.merge(cbsa_crosswalk, on="FIPS", how="left")

manual_agg = {
    "2010 Population": "sum"
}

ignore_list = [
    "FIPS",
    "Category"
]

categorical_columns = ['FIPS', 'cbsacode', 'csacode', 'cbsatitle', 'csatitle']
metro_df[categorical_columns] = metro_df[categorical_columns].astype('category')

# Add a default aggregation method (e.g., "max") for all other columns
default_agg = {col: "mean" for col in metro_df.columns if col not in manual_agg and col not in ignore_list}
agg_methods = {**manual_agg, **default_agg}

# Group by MetroArea and Category
aggregated_data = metro_df.groupby(["cbsacode", "Category"]).agg(agg_methods).reset_index()
aggregated_data
```
